```python
import cv2  #opencv library
import matplotlib.pyplot as plt  # Matplotlib used for visualization
%matplotlib inline
```

```python
img1=cv2.imread(r'C:\Users\HP\Desktop\doc.jfif') #Reading the jfif formmatted image from its path to img1
img2=cv2.imread(r'C:\Users\HP\Desktop\hp1.jpg')  #Reading the jpg formatted image from its path to img2
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB) #Converting from BGR(blue,green,red) color space to RGB using cvtColor
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1) #Displaying the image
```

```python
flip_img=cv2.flip(img1,1) # '1' Flips image to left side, '0' to upside down and '-1' to combination of both 0 and 1
```

```python
plt.imshow(flip_img)  #Showing flipped image
```

```python
plt.imshow(img2) #Displaying Image2
```

```python
img1.shape #Gives the shape (height,width, no.of color channels(RGB)) of image1
```

```python
img2.shape #Gives the shape (height,width, no.of color channels(RGB)) of image2
```

```python
img1=cv2.resize(img1,(2560,1600)) #Both images should be of same size to blend. 
```

```python
img2=cv2.resize(img2,(2560,1600))
```

```python
blended=cv2.addWeighted(src1=img1,alpha=0.8,src2=img2,beta=0.6,gamma=0) #Increase in alpha increases img1 weightage,beta for img2
```

```python
plt.imshow(blended)
```

```python

cv2.imwrite('blended.jpg',blended) #Saves the blended img under the name blended.jpg 
```
