```python
import cv2  #opencv library
import matplotlib.pyplot as plt  
          # Matplotlib used for visualization
%matplotlib inline
import numpy as np
```

```python
img=cv2.imread(r'C:\Users\HP\Desktop\Blog\Opencv\h.jpg')
```

```python
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
kernel=np.ones((5,5),np.float)/25
```

```python
plt.imshow(cv2.blur(img,ksize=(5,5))) #Default Blurring Technique
```

```python
plt.imshow(cv2.filter2D(img,-1,kernel)) #For Smoothing
```

```python
plt.imshow(cv2.GaussianBlur(img,(5,5),0)) #Using Gaussian Technique
```

```python
plt.imshow(cv2.medianBlur(img,5))          #Reduces noise
```

```python
plt.imshow(cv2.bilateralFilter(img,9,75,75))   #Reduces unwanted noise & sharp edges but slow
```
